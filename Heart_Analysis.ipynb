## Attributes with respect to target value

```python
#kde: This is a boolean parameter that determines whether to plot the Kernel Density Estimate (KDE) along with the histogram.
#If kde=True, a smoothed curve representing the estimated probability density function will be overlaid on the histogram.

#hue: This parameter allows you to introduce a categorical variable to color the data points in the plot.
#the 'target' column is used as the hue variable, meaning that the histogram bars or KDE curves will be colored differently based on the categories
#present in the 'target' variable. This is useful for comparing the distributions of 'thalach' for different categories.

plt.figure(figsize=(10,6))
p = sns.histplot(data = df , x="thalach" , kde = True , hue ='target')
plt.title("Maximum heart rate " , fontsize=20)
plt.savefig('C:/Users/Bhavya Parekh/Desktop/Graphs_SDP/Heart_Analysis/max_pulse_rate.png', format='png')
plt.show(p)
```

```python

#The term "thal" often refers to the thallium stress test, a diagnostic test used in nuclear medicine to evaluate blood flow to the heart. 
#Thallium is a radioactive substance that is injected into the bloodstream, and a special camera is used to take images of the heart muscle.
plt.figure(figsize=(10,6))
q = sns.histplot(data = df , x="thal" , kde = True , hue ='target')
plt.title("Thalassemia" , fontsize=20)
plt.savefig('C:/Users/Bhavya Parekh/Desktop/Graphs_SDP/Heart_Analysis/img4.png', format='png')
plt.show(q)
```

```python
#palette: It is the parameter that determines the color palette to be used for coloring different levels in the 'hue' variable.
#In your case, 'bright' is the specified palette. Seaborn provides various built-in color palettes, and 'bright' is one of them.
#It typically consists of a set of distinct, vibrant colors that are well-suited for categorical data.
#When you set the palette to "bright" in Seaborn, it assigns a different bright color to each category or level in the categorical variable. 
#This can enhance the visual appeal of your plots and make it easier for viewers to interpret and compare the different groups.
plt.figure(figsize=(9,5))
p=sns.countplot(x='sex' , data=df,hue='target',palette='bright')
plt.xlabel('Gender',fontsize=20)
plt.title('Gender based on heart disease' , fontsize=22)
plt.savefig('C:/Users/Bhavya Parekh/Desktop/Graphs_SDP/Heart_Analysis/img5.png', format='png')
plt.show(p)
# 0 is female and 1 is male
```

```python
plt.figure(figsize=(9,5))
p=sns.relplot(data = df , x='thal' , y='trestbps' , hue='target',size='chol',palette='muted')# muted used for subdued or less vibrant
#the size parameter is often used to specify the size of markers or elements in a plot
plt.xlabel('Thalessemia',fontsize=20)
plt.ylabel('Resting bp',fontsize=20)
plt.title(' Thalessemia vs RestingBp',fontsize=24)
plt.savefig('C:/Users/Bhavya Parekh/Desktop/Graphs_SDP/Heart_Analysis/img12.png', format='png')
plt.show(p)
```

```python
plt.figure(figsize=(9,5))
p=sns.jointplot(data = df , x='thal' , y='oldpeak' ,hue='target' , kind='kde')
#kind: This parameter determines the type of plot to be drawn. In your example, kind='kde' specifies a Kernel Density Estimate (KDE) plot. 
#A KDE plot provides a smoothed representation of the joint distribution of the two variables, showing contours of higher density.
# joint plot of the variables 'thal' and 'oldpeak', using a Kernel Density Estimate (KDE) plot to visualize the distribution of data points. 
p.plot_joint(sns.kdeplot,color='r')
#This method is used to plot the joint distribution of two variables.
#sns.kdeplot: This specifies the type of plot to be used for the joint distribution.it's a Kernel Density Estimate (KDE) plot.
p.set_axis_labels('Thalessemia','oldpeak',fontsize=20)
#This method is used to set labels for the x and y axes.
p.fig.suptitle('thalessemia vs oldpeak' ,fontsize=20)
#This method is used to set the overall title for the entire figure.
plt.savefig('C:/Users/Bhavya Parekh/Desktop/Graphs_SDP/Heart_Analysis/img13.png', format='png')
plt.show(p)
```
